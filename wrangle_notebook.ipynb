# `wrangle.py` Functions
- `get_data()`
- `clean_data()`
- `write_data()`

```python
def get_data(url):
    '''
    This function takes in the wikipedia url and writes the table data from that wikitable on the page
    to the initial data csv (this will be written over with the cleaned data in the final function.
    
    It returns the table data in a pandas df.)
    '''
    # Working out code to put into function
    html = urlopen(url)

    # making soup
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.findAll('table', {'class':'wikitable'})[0]
    rows = table.findAll('tr')
    print('Soup is done.')
    
    # writing each row to a file
    with open('returnData.csv', 'wt+', newline = '') as f:
        writer = csv.writer(f)
        for row in rows:
            rowList = []
            for cell in row.findAll(['td', 'th']):
                rowList.append(cell.get_text())
            writer.writerow(rowList)
    print('File is saved.')
    
    return pd.read_csv('returnData.csv') 
```

```python
get_data('https://en.wikipedia.org/wiki/S%26P_500#cite_note-sp500r-2')
```

```python
# clean function 
def clean_data(df):
    '''
    This function takes in a df of raw data and returns a clean df after performing the following steps:
        1. Cleaning column names and only returning the columns of interest,
        2. Removing last 4 rows with aggregate data
        3. Removing special characters and casting string values as floats, and
        4. Casting year column back to string value
    '''
    
    # column cleanup: removing line break characterts and replacing spaces with underscores
    df.columns = [col.replace('\n', '').replace(' ', '_').lower() for col in df.columns]
    
    # writing over full df to only include columns of interest
    df = df[['year', 'total_annual_return_including_dividends', 'value_of_$1.00_invested_on_january_1,_1970']]
    
    # renaming columns
    df.rename(columns = {'year':'year', 'total_annual_return_including_dividends':'annual_return',
       'value_of_$1.00_invested_on_january_1,_1970':'value_of_dollar'}, inplace = True)
    
##############################
# checkpoint print statement #
    print('Column cleanup complete.')
##############################
    
    # removing spacial characters $ amd % from df values
    df = df.applymap(lambda x: x.replace('\n', '').replace('$', '').replace('%', '')\
                    .replace('\U00002013', '-')).iloc[:52]
    
    # removing ascii hypthen (won't convert to negative when casting strings and casting string values as floats)
    df = df.applymap(lambda x: re.sub(r'[^\x00-\x7F]+', '-', x)).applymap(lambda x: float(x))
    
    # casting year as string
    df.year = df.applymap(lambda x: str(int(x))).year
##############################
# checkpoint print statement #
    print('DataFrame Cleanup Complete.')
##############################

    return df
```

```python
def wrangle_data(cleanedFileName, rawFileName, url):
    '''
    This function checks to see if existing clean data file exists locally. If it does prints a statements 
    advising this and returns the file written to a pandas df.
    
    If a local file of cleaned data does not exist, it then checks for a local file with the raw data. If the 
    raw data file exists it then cleans the data and writes the cleaned data to csv file, returning the 
    written file to a pandas df..
    
    If no data files exist, it parses the text of a given table from an input url and writes the table to
    a file containing the raw. It then cleans that data and writes the clean data to a file, returning the 
    written file to a pandas df.
    '''
    # step 1: check for clean file
    if os.path.isFile(cleanedFileName):
        print('Cleaned file exists locally. Writing to df.')
        df = pd.read_csv(cleanedFileName)
    else:
        if os.path.isFile(rawFileName):
            print('Raw file exists locally. Cleaning')
            df = clean_data(pd.read_csv(rawFileName))
            df.to_csv('cleanData.csv')
            print('Raw data cleaned and written to csv.')
        else:
            print('Raw data file not found locally. Acquiring data from wikitable.')
            rawData = get_data(url)
            rawData.to_csv('rawData.csv')
            print('Raw data file saved. Cleaning Data')
            df = clean_data(rawData)
            df.to_csv('cleanData.csv')
            print('Data aquisition and cleaning complete.')
    return df
```
